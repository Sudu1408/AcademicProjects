**REMOVE SPECIAL CHARACTERS FROM DATA**
**AND FORMAT SOME COLUMN DATA**

```python
df.update(df.discount.apply(lambda x: x.split('%')[0]))

#removes % from the values in discount column...
```

```python
def hours(x):
    
    if len(x) > 3:
        val = x.split()
        hours = val[0].split('h')[0]
        minutes = val[1].split('m')[0]
        
        return hours
    
    else :
        
        return 0
    
def minutes(x):
    
    if len(x) > 3:
        val = x.split()
        hours = val[0].split('h')[0]
        minutes = val[1].split('m')[0]
        
        return minutes
    
    else :
        
        return x.split('m')[0]
    
df['hours'] = df.duration.apply(hours)
df['minutes'] = df.duration.apply(minutes)


#splits values in duration and place it in 2 columns (hours,minutes)...
```

```python
df.hours = df.hours.astype('float64')
df.minutes = df.minutes.astype('float64')

df['duration'] = np.nan

df.duration.fillna((df.hours*60)+df.minutes,inplace = True)

df.duration = df.duration.astype('float64')

df.drop('hours',axis = 1, inplace = True)
df.drop('minutes',axis = 1, inplace = True)


#Replace values in duration with total duration in minutes...
#datatype of column 'duration' is changed to 'float64'...
```

```python
df.update(df.last_updated.apply(lambda x: x[-2::]))

#retrieve last characters of string in last updated (year number)...
```

```python
df.discount = df.discount.astype('float64')
df.last_updated = df.last_updated.astype('float64')

#change datatype of specified columns to 'float64'...
```
